```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset,DataLoader
import torchaudio.transforms as T
from torchsummary import summary
import numpy as np
import pandas as pd
import re
import glob
import os 
import sys
import matplotlib.pyplot as plt
from tqdm import tqdm
import seaborn

device = torch.device("cuda:0" if torch.cuda.is_available() else 'cpu')
```

```python
# build the dataset
class MyDataset(Dataset):
    def __init__(self,root_dir):
        self.root_dir=root_dir
        
        self.filenames=sorted(glob.glob(os.path.abspath("{root}/*.csv".format(root=root_dir))))
        self.waveforms=self.getwaveforms(self.filenames)
        self.transform=T.MelSpectrogram(sample_rate=2000,n_fft=256)
        print("convert to vector_arrays...")
        self.vector_arrays=np.log10(self.transform(torch.tensor(self.waveforms).to(torch.float32)).squeeze())
        self.vector_arrays=self.vector_arrays.transpose(1,2)
        maxvalue,_=torch.max(abs(self.vector_arrays),dim=2)
        maxvalue = maxvalue.unsqueeze(2)
        self.vector_arrays=self.vector_arrays/maxvalue
        print("complete!")
        
    def getwaveforms(self,files):
        files_num=len(files)
        waveforms=[]
        print("building the dataset...")
        for i in tqdm(range(files_num)):
            waveforms.append([pd.read_csv(files[i]).iloc[:,0].values])
        #waveforms=torch.tensor(waveforms).to(torch.float32)
        return waveforms
    
    def __len__(self):
        return len(self.filenames)
    
    def __getitem__(self,idx):
        label=re.findall(r"\\\d+-",self.filenames[idx])
        return self.vector_arrays[idx],int(label[0][1:-1])
```

```python
dataset=MyDataset("F:/PycharmData/Jupyter notebook/ACS_AEM_Codes/datasets/石墨烯列csv")
dataset_len=len(dataset)
trainset_len=int(0.6*dataset_len)
trainset,testset=torch.utils.data.random_split(dataset,[trainset_len,dataset_len-trainset_len])
#trainset,testset=dataset,dataset
def pad_sequence(batch):
    batch=[item for item in batch]
    batch=torch.nn.utils.rnn.pad_sequence(batch,batch_first=False,padding_value=0.)
    return batch


def collate_fn(batch):
    tensors,targets=[],[]
    for vector_array,label in batch:
        tensors+=[vector_array]
        targets+=[torch.tensor(label)]
    tensors=pad_sequence(tensors)
    targets=torch.stack(targets)
    return tensors,targets

batch_size=64
trainloader=DataLoader(trainset,batch_size=batch_size,shuffle=True,collate_fn=collate_fn)
testloader=DataLoader(testset,batch_size=batch_size,shuffle=True,collate_fn=collate_fn)
```

```python
# define the GRU model
class GRUmodel(nn.Module):
    def __init__(self,input_size=128,output_size=10):
        super().__init__()
        self.gru=nn.GRU(input_size,hidden_size=128,num_layers=2)
        self.fc1=nn.Linear(128,64)
        self.fc2=nn.Linear(64,output_size)
        
    def forward(self,x,h0):
        x,hn=self.gru(x,h0)
        x=x[-1]
        x=F.relu(self.fc1(x))
        x=F.log_softmax(self.fc2(x),dim=-1)
        return x
    
model=GRUmodel()
model=model.to(device)

print(model)
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

n = count_parameters(model)
print("Number of parameters: %s" % n)
```

```python
# define the optimizer
optimizer = optim.SGD(model.parameters(), lr=0.1, weight_decay=0.0001)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=200, gamma=0.1) 
```

```python
def train(model,epoch,log_interval):
    model.train()
    for batch_idx, (data, target) in enumerate(trainloader):
        data=data.to(device).to(torch.float32)
        target=target.to(device)
        h0=torch.zeros(2,data.size()[1],128).to(device)
        output = model(data,h0)
        loss = F.nll_loss(output.squeeze(), target)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if batch_idx % log_interval == 0:
            print(f"Train Epoch: {epoch} [{batch_idx * len(data)}/{len(trainloader.dataset)} ({100. * batch_idx / len(trainloader):.0f}%)]\tLoss: {loss.item():.6f}")
            losses.append(loss.item())
```

```python
def number_of_correct(pred, target):
    # count number of correct predictions
    return pred.squeeze().eq(target).sum().item()


def get_likely_index(tensor):
    # find most likely label index for each element in the batch
    return tensor.argmax(dim=-1)


def test(model, epoch):
    model.eval()
    correct = 0
    for data, target in testloader:

        data = data.to(torch.float32).to(device)
        target=target.to(device)
        h0=torch.zeros(2,data.size()[1],128).to(device)
        # apply transform and model on whole batch directly on device
        output = model(data,h0)

        pred = get_likely_index(output)
        correct += number_of_correct(pred, target)   
  

    print(f"\nTest Epoch: {epoch}\tAccuracy: {correct}/{len(testloader.dataset)} ({100. * correct / len(testloader.dataset):.0f}%)\n")
    return correct/(len(testloader.dataset))
```

```python
log_interval = 20
n_epoch = 400

losses = []
accuracy=[]

for epoch in tqdm(range(1, n_epoch + 1)):
    train(model, epoch, log_interval)
    accuracy+=[test(model, epoch)]
    scheduler.step()
```

```python
mylosses=np.array(losses)/max(losses)
print(len(mylosses))
print(len(accuracy))
```

```python
plt.figure()
plt.plot(range(len(mylosses)),mylosses)
plt.plot(range(len(accuracy)),accuracy)
plt.show()
```

```python
df=np.array([accuracy,mylosses])
df=df.T
df=pd.DataFrame(df,columns=["acc","loss"])
df.to_csv("result.csv")
```

```python
def onetest(model, epoch):
    model.eval()
    correct = 0
    preds=[]
    targets=[]
    for data, target in testloader:
        targets+=target
        data = data.to(torch.float32).to(device)
        target=target.to(device)
        h0=torch.zeros(2,data.size()[1],128).to(device)
        # apply transform and model on whole batch directly on device
        output = model(data,h0)

        pred = get_likely_index(output)
        preds+=pred
        correct += number_of_correct(pred, target)   
  

    print(f"\nTest Epoch: {epoch}\tAccuracy: {correct}/{len(testloader.dataset)} ({100. * correct / len(testloader.dataset):.0f}%)\n")
    return correct/(len(testloader.dataset)),preds,targets
acc,preds,targets=onetest(model,1)
print(acc)
preds=list(map(lambda x:int(x.cpu()),preds))
targets=list(map(lambda x:int(x),targets))
print(preds)
print(targets)
```
